```python
from tinydb import TinyDB, Query
from collections import defaultdict

# Open the database
db = TinyDB('legal_db.json')

# Get all documents
all_docs = db.all()

# Create a dictionary to store connections
connections = defaultdict(list)

# Function to check if a string is in any field of a document
def string_in_doc(string, doc):
    return any(string.lower() in str(value).lower() for value in doc.values())

# Iterate through all documents
for i, doc in enumerate(all_docs):
    doc_id = doc.get('id', f'doc_{i}')  # Use 'id' field if exists, otherwise create a unique id
    
    # Query for documents that contain the current document's ID
    query = Query()
    matching_docs = db.search(query.fragment({key: doc_id}) for key in doc.keys())
    
    # Add connections
    for other_doc in matching_docs:
        other_doc_id = other_doc.get('id', f'doc_{i}')
        if other_doc_id != doc_id:
            connections[doc_id].append(other_doc_id)

# Print connections
for doc_id, connected_docs in connections.items():
    if connected_docs:
        print(f"Document {doc_id} is referenced by: {', '.join(connected_docs)}")

# Close the database
db.close()
```
